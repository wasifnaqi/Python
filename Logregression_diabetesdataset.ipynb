```python
for col in df.columns:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3-q1

  upper_limit = q3+(1.5*iqr)
  lower_limit = q1-(1.5*iqr)

  df = df[(df[col]<upper_limit) & (df[col]>lower_limit)]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Simulated data: tumor size (cm) and diagnosis (0 = benign, 1 = malignant)
np.random.seed(42)
sizes = np.array([1.2, 1.5, 2.0, 2.3, 2.5, 2.8, 3.0, 3.2, 3.5, 3.8, 4.0, 4.2, 4.5]).reshape(-1, 1)
labels = np.array([0, 0, 0, 0, 1, 0, 1, 1, 1, 1, 1, 1, 1])

# Fit logistic regression model
model = LogisticRegression()
model.fit(sizes, labels)

# Predict probability of being malignant
probabilities = model.predict_proba(sizes)[:, 1]  # Probability of class 1

# Predicted class (using 0.5 threshold)
predictions = model.predict(sizes)

# Plot the results
plt.figure(figsize=(10, 6))

# Scatter plot of actual data
plt.scatter(sizes, labels, color='black', label='Actual data (0=Benign, 1=Malignant)')

# Plot predicted probabilities
plt.plot(sizes, probabilities, color='blue', label='Predicted probability (logistic curve)', linewidth=2)

# Add predicted class (threshold at 0.5)
plt.scatter(sizes, predictions, color='red', marker='x', s=100, label='Predicted class', zorder=5)

# Customize plot
plt.xlabel('Tumor Size (cm)')
plt.ylabel('Probability of Malignant (Class 1)')
plt.title('Logistic Regression: Classification via Probability')
plt.ylim(-0.1, 1.1)
plt.legend()
plt.grid(alpha=0.3)

# Show plot
plt.show()
```

```python
print(f"Intercept: {model.intercept_[0]:.2f}")
print(f"Coefficient (slope): {model.coef_[0][0]:.2f}")
```
